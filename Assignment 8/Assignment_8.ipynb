```python
import matplotlib.pyplot as plt

# Measured execution times (replace with your actual values)
processes = [1, 2, 4, 8]
times = [0.543120, 0.320874, 0.180554, 0.128005]

# Compute speedup
speedup = [times[0] / t for t in times]

# Plot 1: Execution Time vs Number of Processes
plt.figure()
plt.plot(processes, times, marker='o', linewidth=2)
plt.title("Execution Time vs Number of Processes (2D Convolution MPI)")
plt.xlabel("Number of Processes")
plt.ylabel("Execution Time (seconds)")
plt.grid(True)
plt.show()

# Plot 2: Speedup vs Number of Processes
plt.figure()
plt.plot(processes, speedup, marker='o', color='orange', linewidth=2)
plt.title("Speedup vs Number of Processes (2D Convolution MPI)")
plt.xlabel("Number of Processes")
plt.ylabel("Speedup ()")
plt.grid(True)
plt.show()
```
