```python
import matplotlib.pyplot as plt

# ---------------- Word Search Data ----------------
word_seq = 0.00663531
word_par = 0.0218751
word_speedup = 0.303327
word_waste = 0

# ---------------- 8 Puzzle Data ----------------
puzzle_seq = 0.000003078
puzzle_par = 0.000509473
puzzle_speedup = 0.0178753
puzzle_waste = 25

# === BAR CHART: Sequential vs Parallel ===
problems = ['Word Search', '8 Puzzle']
seq_times = [word_seq, puzzle_seq]
par_times = [word_par, puzzle_par]

x = range(len(problems))
plt.figure(figsize=(8,5))
plt.bar(x, seq_times, width=0.4, label='Sequential', color='skyblue')
plt.bar([i + 0.4 for i in x], par_times, width=0.4, label='Parallel', color='orange')
plt.xticks([i + 0.2 for i in x], problems)
plt.ylabel('Time (seconds)')
plt.title('Sequential vs Parallel Execution Time')
plt.legend()
plt.show()

# === LINE CHART: Speedup ===
plt.figure(figsize=(8,5))
plt.plot(problems, [word_speedup, puzzle_speedup], marker='o', color='green')
plt.title('Speedup Comparison')
plt.ylabel('Speedup (x)')
plt.grid(True)
plt.show()

# === BAR CHART: Wasted Computation ===
plt.figure(figsize=(8,5))
plt.bar(problems, [word_waste, puzzle_waste], color='salmon')
plt.title('Wasted Computation (%)')
plt.ylabel('Percentage')
plt.show()
```
